### Function to collect data

```python
# flush any junk left in the serial buffer
arduinoData.flushInput()

# Initialize empty plot lists
FS1_plot = []   # empty array for sensor 0 data
FS2_plot = []   # empty array for sensor 1 data
FS3_plot = []   # empty array for sensor 2 data
FS4_plot = []   # empty array for sensor 3 data
FS5_plot = []   # empty array for sensor 4 data
FS6_plot = []   # empty array for sensor 5 data
FS7_plot = []   # empty array for sensor 6 data
FS8_plot = []   # empty array for sensor 7 data
FS9_plot = []   # empty array for sensor 8 data

#with open(r'testing_raw.csv', 'a') as f: #add.csv
    #writer = csv.writer(f)
with open(r'eskin_raw.csv', 'a') as f1: #add.csv
    writer1 = csv.writer(f1)
    # Continuously Running Loop    
    for i in tqdm(running_list):
        arduinoData.flushInput()
        print(i)
        input('enter data:')   
        j = 0
        line = []
        #linep = []
        while j<90:
            while(arduinoData.inWaiting()==0):  # if there is not data
                pass 
            arduinoData.reset_input_buffer()
            arduinoRow = arduinoData.readline() # for all data read Serial Lines
            arduinoRow = arduinoRow.decode("utf-8")
            arduinoRow = arduinoRow.replace('\r\n','')
            dataArray = arduinoRow.split(',')  # slit data at ' ' defined in C++ program

            #print(dataArray)

            FS1 = float(dataArray[0])   # turn string data into number
            FS2 = float(dataArray[1])   # turn string data into number
            FS3 = float(dataArray[2])   # turn string data into number
            FS4 = float(dataArray[3])   # turn string data into number
            FS5 = float(dataArray[4])   # turn string data into number
            FS6 = float(dataArray[5])   # turn string data into number
            FS7 = float(dataArray[6])   # turn string data into number
            FS8 = float(dataArray[7])   # turn string data into number
            FS9 = float(dataArray[8])   # turn string data into number

            line=line+ [abs(FS1),abs(FS2),abs(FS3),abs(FS4),abs(FS5),
                        abs(FS6),abs(FS7),abs(FS8),abs(FS9)]

            #linep=linep+ [power(FS1),power(FS2),power(FS3),power(FS4),power(FS5),
                       # power(FS6),power(FS7),power(FS8),power(FS9)]

            # keeps graph from getting big, only display 100 values
            cnt += 1
            j+=1

        Fs1 = line[0::9]
        Fs2 = line[1::9]
        Fs3 = line[2::9]
        Fs4 = line[3::9]
        Fs5 = line[4::9]
        Fs6 = line[5::9]
        Fs7 = line[6::9]
        Fs8 = line[7::9]
        Fs9 = line[8::9]

        ls = list(range(90))

        plt.plot(ls,Fs1,linewidth=1.0)
        plt.plot(ls,Fs2,linewidth=1.0)
        plt.plot(ls,Fs3,linewidth=1.0)
        plt.plot(ls,Fs4,linewidth=1.0)
        plt.plot(ls,Fs5,linewidth=1.0)
        plt.plot(ls,Fs6,linewidth=1.0)
        plt.plot(ls,Fs7,linewidth=1.0)
        plt.plot(ls,Fs8,linewidth=1.0)
        plt.plot(ls,Fs9,linewidth=1.0)
        plt.show()

        line.insert(len(line),i)

        keep = input('Is this good data:')
        if keep == '':
            Full.append([line])
            #writer.writerow(line)
            
            writer1.writerow(line)
            if 'r' in i:
                rub += 1
            elif 'p' in i:
                poke += 1
            elif 'l' in i:
                slap += 1
            elif 't' in i:
                stroke += 1
            print('data:',len(line), 'Counts:','p',poke,'r',rub,'sl',slap,'st',stroke)
        else:
            missing.append(i)
            print('added to missing')
        tempres.pop(0)
        print(len(tempres))

print('Complete ...', len(missing),'Data is corrupted')
```
